2. The digital root of a number n is obtained as follows: Add up the digits n to get a new number.
Add up the digits of that to get another new number. Keep doing this until you get a number
that has only one digit. That number is the digital root.
For example, if n= 45893, we add up the digits to get 4 + 5 + 8 + 9 + 3 = 29. We then add up
the digits of 29 to get 2 + 9 = 11. We then add up the digits of 11 to get 1 + 1 = 2. Since 2 has
only one digit, 2 is our digital root.
Write a function that returns the digital root of an integer n. [Note: there is a shorteut, where the digital root is equal to n mod 9, but do not use that here.]

```python
def raíz_digital():
    """La función recibe un número, y sacara su raiz digital 
    (la suma de cada uno de sus digitos hasta que de 1 sola cifra)"""

    x= input("Introduce un número: ")
    estado= True
    while estado==True:
        lista = []
        for i in x:
            lista.append(int(i))
        y=0
        if len(lista)!=1:
            for a in lista:
                y+=a
                x=str(y)
        else:
            print("La raíz digital es:",x)
            estado=False

    return
```
